```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("datasets/Market_Basket_Optimisation.csv",header=None)
df =df.fillna('')
df.head()
```

```python
List_Items=[]
for i in range(7501):
    l1=list()
    l1=list(df.iloc[i])
    List_Items.append(l1)
```

```python
new_list = []
for x in List_Items:
    l1=list()
    for y in x:
        if y != '':
            l1.append(y)
    new_list.append(l1)
```

```python
ItemCount=dict()
for j in range(7501):
    for i in df.columns:
        if (df.iloc[j,i]==''):
            break
        elif (df.iloc[j,i] not in ItemCount):
            ItemCount[df.iloc[j,i]]=1
        else:
            ItemCount[df.iloc[j,i]]=ItemCount[df.iloc[j,i]]+1
```

```python
ItemCount
```

```python
def ItemDicttoFrame(d1):
    d1 = pd.DataFrame.from_dict(d1,orient='index')
    d1.rename(columns={0:'Counts'},inplace=True)
    d1.index.name='Items'
    d1.reset_index(inplace=True)
    d1.sort_values(by=['Counts'],ascending=False,inplace=True,ignore_index=True)
    return d1
```

```python
Item1Counts = ItemDicttoFrame(ItemCount)
Item1Counts
```

```python
Item1Counts['Counts'].describe()
```

```python
Freq1 = Item1Counts[Item1Counts['Counts']>117]
Freq1
```

```python
from itertools import combinations
```

```python
def CandidateItemGeneration(c=2,freq=Freq1):
    d1 = dict()
    List1 = list(combinations(Freq1.Items,c))
    for i in List1:
        d1[i]=0
        for x in new_list:
            count=0
            for j in i:
                if j in x:
                    count=count+1
            if count==c:
                d1[i]=d1[i]+1
    return d1
```

```python
Item2Counts = CandidateItemGeneration()
Item2Counts = ItemDicttoFrame(Item2Counts)
```

```python
Item2Counts['Items'] = Item2Counts['Items'].map(lambda x: str(x).lstrip(r"(").rstrip(r")"))
Item2Counts
```
